### USING FOR LOOP APPEND ALL THE DATA INSIDE LISTS  
#### STATIC 

```python
name=[]
mileage=[]
dealer_name=[]
reviews=[]
ratings=[]
price=[]

for result in res:
    #name
    try:
        name.append(result.find("h2").get_text())
        mileage.append(result.find("div",{"class":"mileage"}).get_text())
        dealer_name.append(result.find("div",{"class":"dealer-name"}).get_text().strip())
        reviews.append(result.find("span",{"class":"sds-rating__link"}).get_text())
        ratings.append(result.find("span",{"class":"sds-rating__count"}).get_text())
        price.append(result.find("span",{"class":"primary-price"}).get_text())
    except:
        name.append("n/a")
        mileage.append("n/a")
        dealer_name.append("n/a")
        reviews.append("n/a")
        ratings.append("n/a")
        price.append("n/a")
        
    
```

```python
reviews
```

```python
data=pd.DataFrame({"Car Name":name,"Mileage":mileage,"Dealer Name":dealer_name,"Ratings":ratings,"Rating Count":reviews,"Price":price})
```

```python
data.head()
```

```python
data["Rating Count"]=data["Rating Count"].apply(lambda x :x.strip('reviews').strip('()'))
```

### SAVE TO CSV OR ANY OTHER FORMAT

### SCRAPING ENTIRE MULTIPLE WEB PAGES
#### DYNAMIC

```python
name=[]
mileage=[]
dealer_name=[]
reviews=[]
ratings=[]
price=[]

for i in range(1,11):
    website="https://www.cars.com/shopping/results/?page="+str(i)+"&page_size=20&dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=dodge&maximum_distance=20&mileage_max=&sort=best_match_desc&stock_type=all&year_max=&year_min=&zip="
    request=requests.get(website)
    soup=bs4(request.content,"html.parser")
    res=soup.find_all("div",{"class":"vehicle-card"})
    for result in res:
        try:
            name.append(result.find("h2").get_text())
        except:
            name.append("n/a")
        try:
            mileage.append(result.find("div",{"class":"mileage"}).get_text())
        except:
            mileage.append("n/a")
        try:
            dealer_name.append(result.find("div",{"class":"dealer-name"}).get_text().strip())
        except:
            dealer_name.append("n/a")
        try:
            ratings.append(result.find("span",{"class":"sds-rating__count"}).get_text())
        except:
            ratings.append("n/a")
        try:
            price.append(result.find("span",{"class":"primary-price"}).get_text())
        except:
            price.append("n/a")
        try:
            reviews.append(result.find("span",{"class":"sds-rating__link"}).get_text())
        except:
            reviews.append("na")
        
       
```

```python
data=pd.DataFrame({"Car Name":name,"Mileage":mileage,"Dealer Name":dealer_name,"Ratings":ratings,"Rating Count":reviews,"Price":price})
```
